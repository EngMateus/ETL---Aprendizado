```python
import pandas as pd
import requests
from io import StringIO
import numpy as np
```

```python
# URL do arquivo CSV
url = "https://dadosabertos.aneel.gov.br/dataset/da3d8da5-bdc0-471a-b57e-09e49f557c9a/resource/eca732ac-4f08-401f-b1cf-f9eff88a9b20/download/sigec-taxa-de-fiscalizacao.csv"
```

```python
# Realizar o download do arquivo CSV
response = requests.get(url)
#### 8859 -> Portugues BR
response_content = response.content.decode('ISO-8859-1')
```

```python
# Carregar os dados em um DataFrame
data = StringIO(response_content)
sigec = pd.read_csv(data, sep=';', encoding='latin1',decimal=',')
```
